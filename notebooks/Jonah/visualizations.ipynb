```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Increasing end date of catalog sequences (making sequences longer and longer)
end_date = {}
end_date['365'] = -1.4847609996795654
end_date['2*365'] = -1.4712773561477661
end_date['4*365'] = -1.4615024328231812
end_date['8*365'] =  -1.4681977033615112
end_date['16*365'] = -2.0318117141723633
years = list(['365', '2*365', '4*365', '8*365', '16*365' ])
loss = list(end_date.values())
plt.plot(years, loss)
plt.xlabel('end_date parameter')
plt.ylabel('NLL Loss')
plt.title('NLL loss as length of sequences is increased')
```
